```python
import numpy as np
from scipy.stats import t
import matplotlib.pyplot as plt
from scipy import stats


# Parametrik
np.random.seed(123)
n = 10
x = np.random.normal(loc=0, scale=1, size=n)
x_bar = np.mean(x)
var_x = np.var(x)
sd_x = np.std(x)
B = 100
x_bar_boot = np.zeros(B)
for i in range(B):
    x_boot = np.random.choice(x, size=n, replace=True)
    x_bar_boot[i] = np.mean(x_boot)

_ = plt.hist(x_bar_boot)
_ = plt.show()
_ = stats.probplot(x_bar_boot, plot=plt)

# Nonparametrik
x = np.random.uniform(low=1, high=3, size=20)
n = len(x)
x_bar = np.mean(x)
var_x = np.var(x)
sd_x = np.std(x)
var_x_bar = var_x/n
sd_x_bar = np.sqrt(var_x_bar)
B = 100
x_bar_boot = np.zeros(B)
for i in range(B):
    x_boot = np.random.choice(x, size=n, replace=True)
    x_bar_boot[i] = np.mean(x_boot)

_ = plt.hist(x_bar_boot)
_ = plt.show()
_ = stats.probplot(x_bar_boot, plot=plt)

# Bias dan Varians Statistik
x_bar_boot_bar = np.mean(x_bar_boot)
bias = x_bar_boot_bar - x_bar
var_x_bar_boot = np.var(x_bar_boot)
sd_x_bar_boot = np.sqrt(var_x_bar_boot)

# CI Xbar boot bawah
x_bar_boot_bar = np.mean(x_bar_boot)
bias = x_bar_boot_bar - x_bar
var_x_bar_boot = np.var(x_bar_boot)
sd_x_bar_boot = np.sqrt(var_x_bar_boot)

z = t.ppf(0.975, n-1)

# CI bawah
CI_bawah = x_bar_boot_bar - z*(sd_x_bar_boot/np.sqrt(n))
print(CI_bawah)

# CI Atas
CI_atas = x_bar_boot_bar + z*(sd_x_bar_boot/np.sqrt(n))
print(CI_atas)
```
